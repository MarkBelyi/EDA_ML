### Read the Data

```python
data = pd.read_csv('salary.csv')
data_clean = data.dropna(subset=['Age', 'Salary'])
```

#### Set a size of Figure

```python
plt.figure(figsize=(15, 15))

data_clean[['Years of Experience', 'Salary']].corr()
```

#### What i did:
##### 1. Hexagon Binning: Displays the relationship between years of experience and salary.
##### 2. Contour Plot: Shows the density of salary distribution by age.
##### 3. Violin Plot: Illustrates the distribution of salaries across different education levels.

```python
plt.hexbin(data_clean['Years of Experience'], data_clean['Salary'], gridsize=30, cmap='winter')
plt.colorbar(label='Counts')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Hexagon Binning: Years of Experience vs. Salary')
```

#### Hexagon Binning: Years of Experience vs. Salary:

##### Analysis: The graph shows that the salary increases with increasing experience. The highest density is observed in people with 5-10 years of experience and an average salary.
##### Conclusion: Experience directly affects salary, but the increase after 15 years of experience is not so pronounced.

```python
sns.kdeplot(x=data_clean['Age'], y=data_clean['Salary'], cmap='Reds', fill=False)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Contour Plot: Age vs. Salary')
```

```python
sns.kdeplot(x=data_clean['Age'], y=data_clean['Salary'], cmap='Reds', fill=False, thresh=0)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Contour Plot: Age vs. Salary')
```

#### Contour Plot: Age vs. Salary:

##### Analysis: Density shows that most employees receive salaries between 60k and 120k between the ages of 30 and 50.
##### Conclusion: The average age range (30-50 years) it has the highest income density.

```python
sns.violinplot(x='Education Level', y='Salary', data=data_clean)
plt.xlabel('Education Level')
plt.xticks(rotation=45)
plt.ylabel('Salary')
plt.title('Violin Plot: Salary Distribution Across Education Levels')
plt.tight_layout()
```

#### Violin Plot: Salary Distribution Across Education Levels:

##### Analysis: It can be seen that employees with higher education (master's and PhD) have higher salaries, and with a bachelor's degree there is a wider spread.
##### Conclusion: A higher level of education is associated with higher salaries.

## 1. Bootstrapping (Bootstrap):
#### What it is: A random sampling method with a return for estimating statistical parameters (for example, mean, variance, etc.).
#### Why it is needed: It is used to estimate the uncertainty of the model or to build confidence intervals.
## 2. Bagging (Bootstrap Aggregating):
#### What it is: A method that uses multiple bootstrap samples to train multiple models and then combines their predictions. This helps to reduce the variance and improve the stability of the model.
#### Why it is needed: It is often used in ensemble methods, such as random forests, to improve the accuracy of predictions.

```python
def bootstrap_sampling(data, n_iterations=1000, sample_size=100):
    means = []
    for _ in range(n_iterations):
        sample = data.sample(n=sample_size, replace=True)
        means.append(sample['Salary'].mean())
    return means
```

```python
bootstrap_means = bootstrap_sampling(data_clean)
```

```python
mean_bootstrap = np.mean(bootstrap_means)
mean_bootstrap
```
