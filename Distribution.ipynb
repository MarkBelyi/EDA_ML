```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define ranges for the x values
x_normal = np.linspace(-4, 4, 1000)
x_f = np.linspace(0, 5, 1000)
x_chi2 = np.linspace(0, 10, 1000)
x_t = np.linspace(-4, 4, 1000)
x_beta = np.linspace(0, 1, 1000)
x_exp = np.linspace(0, 5, 1000)
x_poisson = np.arange(0, 20)
x_gamma = np.linspace(0, 20, 1000)

# Subplots for each distribution
fig, axs = plt.subplots(4, 2, figsize=(15, 20))

# Normal distribution
axs[0, 0].plot(x_normal, stats.norm.pdf(x_normal), label='Normal Distribution')
axs[0, 0].set_title('Normal Distribution')
axs[0, 0].legend()

# F-distribution
axs[0, 1].plot(x_f, stats.f.pdf(x_f, dfn=5, dfd=2), label='F-distribution (dfn=5, dfd=2)', color='orange')
axs[0, 1].set_title('F-distribution')
axs[0, 1].legend()

# Chi-square distribution
axs[1, 0].plot(x_chi2, stats.chi2.pdf(x_chi2, df=2), label='Chi-square (df=2)', color='green')
axs[1, 0].set_title('Chi-square Distribution')
axs[1, 0].legend()

# t-distribution
axs[1, 1].plot(x_t, stats.t.pdf(x_t, df=10), label='t-distribution (df=10)', color='red')
axs[1, 1].set_title('t-distribution')
axs[1, 1].legend()

# Beta distribution
axs[2, 0].plot(x_beta, stats.beta.pdf(x_beta, a=2, b=5), label='Beta distribution (a=2, b=5)', color='purple')
axs[2, 0].set_title('Beta Distribution')
axs[2, 0].legend()

# Exponential distribution
axs[2, 1].plot(x_exp, stats.expon.pdf(x_exp, scale=1), label='Exponential Distribution', color='brown')
axs[2, 1].set_title('Exponential Distribution')
axs[2, 1].legend()

# Poisson distribution
axs[3, 0].stem(x_poisson, stats.poisson.pmf(x_poisson, mu=5), label='Poisson Distribution (mu=5)', linefmt='blue', markerfmt='bo')
axs[3, 0].set_title('Poisson Distribution')
axs[3, 0].legend()

# Gamma distribution
axs[3, 1].plot(x_gamma, stats.gamma.pdf(x_gamma, a=2), label='Gamma Distribution (a=2)', color='cyan')
axs[3, 1].set_title('Gamma Distribution')
axs[3, 1].legend()

plt.tight_layout()
plt.show()
```
